# Data-方法-04 贝叶斯学习
- 对西瓜数据集2.0（周志华老师的《机器学习》p76），实现了简单的朴素贝叶斯分类器。

## 1. 读入数据
- 与之前一样，不做更多阐述了

```python
# coding: utf-8

import pandas as pd
# 读入西瓜数据集

path = '../数据集/西瓜数据/西瓜2.csv'

datas = pd.read_csv(path)
datas
```

## 2. 计算先验和后验概率

```python
'''获取数据集中的属性特征'''
def getFeatures(datas):
    features = {}
    
    for feature in datas.keys():
        if(feature == '编号'): continue
        if(feature == '好瓜'):
            features[feature] = {'是':0, '否':0}
            continue
        
        values = {}
        for value in datas[feature]:
            if value not in values.keys():
                values[value] = {'是':0, '否':0}
            
        features[feature] = values
    
    return features
    
features = getFeatures(datas)
features
```

```python
'''计数'''
def countFeatures(datas, features):
    
    for feature in datas.keys():
        if(feature == '编号'): continue
        if(feature == '好瓜'):
            for i in range(len(datas)):
                features[feature][datas['好瓜'][i]] += 1
            continue
        
        for i in range(len(datas)):
            features[feature][datas[feature][i]][datas['好瓜'][i]] += 1
    
    return features

counts = countFeatures(datas, features)
counts
```

```python
'''极大似然估计-计算先验概率和条件概率'''
from copy import deepcopy

def calMLE(counts):
    proMLE = deepcopy(counts)
    
    for key1 in counts.keys():
        if key1=='好瓜':
            proMLE[key1]['是'] = counts[key1]['是']/\
                        (counts[key1]['是']+counts[key1]['否'])
            proMLE[key1]['否'] = proMLE[key1]['否']/\
                        (counts[key1]['是']+counts[key1]['否'])
            continue
        
        for key2 in counts[key1].keys():
            proMLE[key1][key2]['是'] = counts[key1][key2]['是']/\
                        (counts[key1][key2]['是']+counts[key1][key2]['否'])
            proMLE[key1][key2]['否'] = counts[key1][key2]['否']/\
                        (counts[key1][key2]['是']+counts[key1][key2]['否'])
            
    return proMLE

proMLE = calMLE(counts)
proMLE
```

```python
'''贝叶斯估计'''
def calBayes(counts, lam):
    proBayes = deepcopy(counts)
    
    for key1 in counts.keys():
        if key1=='好瓜':
            proBayes[key1]['是'] = (counts[key1]['是']+lam)/\
                        (counts[key1]['是']+counts[key1]['否']+lam*len(counts.keys()))
            proBayes[key1]['否'] = (proMLE[key1]['否']+lam)/\
                        (counts[key1]['是']+counts[key1]['否']+lam*len(counts.keys()))
            continue
        
        for key2 in counts[key1].keys():
            proBayes[key1][key2]['是'] = (counts[key1][key2]['是']+lam)/\
                        (counts[key1][key2]['是']+counts[key1][key2]['否']+\
                        lam*len(counts[key1].keys()))
            proBayes[key1][key2]['否'] = (counts[key1][key2]['否']+lam)/\
                        (counts[key1][key2]['是']+counts[key1][key2]['否']+\
                        lam*len(counts[key1].keys()))
            
    return proBayes
```

```python
# lambda=0 等价于 极大似然估计
proBayes_0 = calBayes(counts, 0)
proBayes_0
```

```python
# lambda=1 拉普拉斯平滑
proBayes_1 = calBayes(counts, 1)
proBayes_1
```

## 3. 预测

```python
'''贝叶斯学习'''
def bayesPredict(pro, features):
    p_yes = 1.0
    p_no = 1.0
    
    for feature in features.keys():
        p_yes *= pro[feature][features[feature]]['是']
        p_no *= pro[feature][features[feature]]['否']
    
    print("正负概率")
    print('- 是: %4f\n- 否: %4f\n'%(p_yes,p_no))
    
    print('预测结果：是' if p_yes>p_no else '预测结果：否')
    
    return None
```

```python
# 预测参数一：
print("预测：色泽=青绿; 根蒂=蜷缩")
pre_1 = {'色泽':'青绿','根蒂':'蜷缩'}

print("\n1. 极大似然估计")
bayesPredict(proBayes_0, pre_1)

print("\n2. 拉普拉斯平滑")
bayesPredict(proBayes_1, pre_1)
```

```python
# 预测参数：
print("预测：色泽=青绿; 根蒂=蜷缩; 脐部=平坦")
pre_1 = {'色泽':'青绿','根蒂':'蜷缩','脐部':'平坦'}

print("\n1. 极大似然估计")
bayesPredict(proBayes_0, pre_1)

print("\n2. 拉普拉斯平滑")
bayesPredict(proBayes_1, pre_1)
```

```python
# 预测参数：
print("预测：色泽=青绿; 根蒂=蜷缩; 脐部=平坦; 纹理=稍糊")
pre_1 = {'色泽':'青绿','脐部':'平坦','纹理':'稍糊'}

print("\n1. 极大似然估计")
bayesPredict(proBayes_0, pre_1)

print("\n2. 拉普拉斯平滑")
bayesPredict(proBayes_1, pre_1)
```

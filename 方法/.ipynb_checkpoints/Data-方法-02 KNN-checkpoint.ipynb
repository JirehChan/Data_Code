# Data-方法-02 KNN


## 1. IRIS 数据集
- IRIS鸢尾花数据集是机器学习领域常用的数据集，包含3类共150条记录，没类各50个数据，每条记录有4个特征（花萼长度、花萼宽度、花瓣长度、花瓣宽度）。
- 我们使用sklearn包，来使用iris数据集，conda安装命令如下,`conda install -n data_env_37 scikit-learn`

```python
'''导入iris数据集'''
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

由输出可以知道，iris中包含：
- 'data': 记录属性
- 'target': 记录对应结果
- 'target_names': 记录结果对应的名称
    (0:'setosa'; 1:'versicolor'; 2:'virginica')
- 'Number of Instances': 150 (每个类50个)

其他还有些信息这里不做解释了。

```python
'''转换iris数据集为DataFrame'''
import pandas as pd

iris_list = []
for i in range(0,150):
    iris_list.append([iris['data'][i],iris['target_names'][iris['target'][i]]]) 
iris_data = pd.DataFrame(iris_list, columns = ['data', 'target'])

iris_data
```

## 2. KNN的线性扫描实现

```python
'''距离度量-Lp距离'''
def calLpDistance(sourcePoint, targetPoint, p=2):
    length = len(targetPoint)
    sum = 0.0
    for i in range(length):
        sum += abs(sourcePoint[i] - targetPoint[i])**p
    sum = sum ** (1/p)
    return sum
```

```python
'''搜索、寻找k近邻'''
def findKNN_linear(sourcePoints, targetPoint, k=1, p=2):
    result = []
    
    for i in range(len(sourcePoints)):
        distance = calLpDistance(sourcePoints['data'][i], targetPoint, p)
        result.append([sourcePoints['data'][i], sourcePoints['target'][i], distance])
    
    result.sort(key = lambda elem: elem[2])
    return pd.DataFrame(result[:k], columns = ['data', 'target', 'distance'])
```

```python
'''决策规则-投票法'''
def voting(data):
    targets = data['target']
    
    target_count = {}
    for i in targets:
        if i not in target_count.keys():
            target_count[i] = 1
        else: target_count[i] += 1    

    return max(target_count,key=target_count.get)
```

```python
'''开始计算'''
target = eval(input('输入目标属性，如(1.0, 2.0, 3.0, 4.0):'))
k = eval(input('k:'))
p = eval(input('Lp的p:'))

result = findKNN_linear(iris_data, target, k, p)
result
```

```python
print('预测结果：', voting(result))
```

## 3. 最近邻法的kd树实现
- 此代码大部分参考 [K近邻法之kd树及其Python实现](https://blog.csdn.net/zkk12345/article/details/76404219)

```python
'''更改数据格式'''
trainingSet = iris_data.values.tolist()
trainingSet
```

```python
'''创建kd树'''
def createTree(dataSet, layer = 0, feature = 2):
    length = len(dataSet)
    dataSetCopy = dataSet[:]
    featureNum = layer % feature
    dataSetCopy.sort(key = lambda x:x[0][featureNum])
    layer += 1
    if length == 0:
        return None
    elif length == 1:
        return {'Value':tuple(dataSet[0][0]), 'Target':dataSet[0][1], 'Layer':layer, 
                'feature':featureNum, 'Left':None, 'Right':None}
    elif length != 1:
        midNum = length // 2
        dataSetLeft = dataSetCopy[:midNum]
        dataSetRight = dataSetCopy[midNum+1:]
        return {'Value':tuple(dataSetCopy[midNum][0]), 'Target':dataSetCopy[midNum][1], 'Layer':layer, 
                'feature':featureNum, 'Left':createTree(dataSetLeft, layer), 'Right':createTree(dataSetRight, layer)}

kdTree = createTree(trainingSet,0,4)
kdTree
```

```python
'''深度搜索，用于构造初始路径'''
def dfs(kdTree, target, tracklist = []):
    tracklistCopy = tracklist[:]
    if not kdTree:
        return None, tracklistCopy
    elif not kdTree['Left']:
        tracklistCopy.append({'Value':kdTree['Value'], 'Target':kdTree['Target']})
        return kdTree['Value'], tracklistCopy
    elif kdTree['Left']:
        pointValue = kdTree['Value']
        feature = kdTree['feature']
        tracklistCopy.append({'Value':kdTree['Value'], 'Target':kdTree['Target']})

        if target[feature] <= pointValue[feature]:
            return dfs(kdTree['Left'], target, tracklistCopy)
        elif target[feature] > pointValue[feature]:
            return dfs(kdTree['Right'], target, tracklistCopy)
        
    return None

target = eval(input('输入目标属性，如(1.0, 2.0, 3.0, 4.0):'))
value, trackList = dfs(kdTree, target)
trackList
```

```python
from math import sqrt

'''计算距离'''
def calDistance(sourcePoint, targetPoint):
    length = len(targetPoint)
    sum = 0.0
    for i in range(length):
        sum += (sourcePoint[i] - targetPoint[i])**2
    sum = sqrt(sum)
    return sum
 
'''寻找节点'''
def findPoint(Tree, value):
    if  Tree !=None and Tree['Value'] == value['Value']:
        return Tree
    elif Tree != None:
        if Tree['Left'] != None:
            return findPoint(Tree['Left'], value) or findPoint(Tree['Right'], value)

'''kd树最近邻搜索'''
def kdTreeSearch(tracklist, target , usedPoint = [] , minDistance = float('inf'), minDistancePoint = None):
    tracklistCopy = tracklist[:]
    usedPointCopy = usedPoint[:]
    
    if not minDistancePoint:
        minDistancePoint = tracklistCopy[-1]
 
    if len(tracklistCopy) == 1:
        return minDistancePoint
    else:
        point = findPoint(kdTree, tracklist[-1])
        
        if calDistance(point['Value'], target) < minDistance:
            minDistance = calDistance(point['Value'], target)
            minDistancePoint = {'Value':point['Value'], 'Target':point['Target']}
        fatherPoint = findPoint(kdTree, tracklistCopy[-2])
        fatherPointval = fatherPoint['Value']
        fatherPointfea = fatherPoint['feature']
 
        if calDistance(fatherPoint['Value'], target) < minDistance:
            minDistance = calDistance(fatherPoint['Value'], target)
            minDistancePoint = {'Value':fatherPoint['Value'], 'Target':fatherPoint['Target']}
 
        if point == fatherPoint['Left']:
            anotherPoint = fatherPoint['Right']
        elif point == fatherPoint['Right']:
            anotherPoint = fatherPoint['Left']
 
        if (anotherPoint == None or {'Value':anotherPoint['Value'], 'Target':anotherPoint['Target']} in usedPointCopy or
            abs(fatherPointval[fatherPointfea] - target[fatherPointfea]) > minDistance):
            usedPoint = tracklistCopy.pop()
            usedPointCopy.append(usedPoint)
            return kdTreeSearch(tracklistCopy, target, usedPointCopy, minDistance, minDistancePoint)
        else:
            usedPoint = tracklistCopy.pop()
            usedPointCopy.append(usedPoint)
            subvalue, subtrackList = dfs(anotherPoint, target)
            tracklistCopy.extend(subtrackList)
            return kdTreeSearch(tracklistCopy, target, usedPointCopy, minDistance, minDistancePoint)
        
    return
 

nnPoint = kdTreeSearch(trackList, target)
nnPoint
```

```python
print('预测结果：', nnPoint['Target'])
```

## 4. 比较一下用时

```python
from time import perf_counter 

target = eval(input('输入目标属性，如(1.0, 2.0, 3.0, 4.0):'))

start = perf_counter()
result = findKNN_linear(iris_data, target, 1, 2)
elapsed = (perf_counter()-start)
print('线性扫描 用时：',elapsed,'\n')

start = perf_counter()
kdTree = createTree(trainingSet,0,4)
elapsed = (perf_counter()-start)
print('kd树创建 用时：',elapsed)

start = perf_counter()
value, trackList = dfs(kdTree, target)
nnPoint = kdTreeSearch(trackList, target)
elapsed = (perf_counter()-start)
print('kd树搜索 用时：',elapsed)
```

- 可以发现，对iris数据集的最近邻算法而言，kd树的实现比线性扫描快了近10倍。

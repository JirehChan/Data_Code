# Data-可视化-03 matplotlib基本图表绘制

```python
import matplotlib.pyplot as plt

# 设置，使中文能正常显示
plt.rcParams['font.sans-serif'] = ['Heiti TC']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

## 1. 柱状图

```python
# 数据
x1 = [5,8,11]
y1 = [5,2,20]
x2 = [6,9,12]
y2 = [3,4,10]

plt.suptitle('柱状图(Bar Chart)')
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=1, hspace=1)

plt.subplot(2,  1,  1)
plt.title('并列柱状图')
plt.bar(x1,y1,color='b',align='center')
plt.bar(x2,y2,color='g',align='center')
plt.legend(['指标1','指标2'])

plt.subplot(2,1,2)
plt.title('叠加柱状图')
plt.bar(x1, y1, width=0.5, color='b', align='center')
plt.bar(x1, y2, bottom=y1, width=0.5, color='g', align='center')
plt.legend(['指标1','指标2'])

plt.savefig('03-01.png')
plt.show()
```

## 2.条状图

```python
# 数据
count = [50,20,30,10]

plt.title('条状图(Bar Chart)')
plt.xlabel('人数')
plt.yticks(range(4),['数字媒体','大数据','嵌入式','移动开发'])

plt.barh(range(4), count, align='center', color='b', alpha=0.5)

plt.savefig('03-02.png')
plt.show()
```

## 3. 饼状图

```python
# 数据
labels = ['数字媒体','大数据','嵌入式','移动开发']
count = [50,20,30,10]

plt.title('饼状图(Pie Chart)')

plt.pie(count, labels=labels, autopct='%1.1f%%')

plt.savefig('03-03.png')
plt.show()
```

## 4. 直方图

```python
# 数据
data = [170,171,171,171.5,172,172,172,172.5,173.5,174,175,175.5,176,177,177.5,178,180]

plt.title('直方图(Histogram)')
plt.xlabel('身高')
plt.ylabel('人数')

plt.hist(data, color='g', alpha=0.5)

plt.savefig('03-04.png')
plt.show()
```

## 5. 散点图

```python
# 数据
x = [1,1,3,6,3,1,6,7,5,8,9,4,9,10,20,13,17,16,15,10,16,11,18]
y = [20,13,17,16,1,1,3,6,3,1,6,7,5,8,9,4,9,10,19,14,15,13,12]

plt.title('散点图(Scatter)')

plt.scatter(x, y, marker='o', color='y', alpha=0.8)

plt.savefig('03-05.png')
plt.show()
```

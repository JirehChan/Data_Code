# Data-可视化-02 matplotlib基本操作

```python
import numpy as np
import matplotlib.pyplot as plt
```

**1. 处理中文显示问题**
- 先查看matplotlib所有包含的字体，找到中文字体'Heiti TC'。
- 然后设置字体即可解决中文显示问题，如果字体库中没有中文字体，可以查看响应教程。

```python
import matplotlib.font_manager

# 显示全部字体
a = sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist])
for i in a:
    print(i)

# 设置，使中文能正常显示
plt.rcParams['font.sans-serif'] = ['Heiti TC']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

**2. 简单示例**
- 包含标题、坐标轴、注释、图例等设置，以及保存图片。

```python
# 绘制的图线
x = np.arange(1,11)
y = 2*x+5

# 显示、设置网格
#plt.grid()
plt.grid(color='k', linewidth='1', linestyle='-.')

# 标题、坐标轴文字
plt.title("标题")
plt.xlabel("x坐标轴")
plt.ylabel("y坐标轴")

# 坐标轴范围
#plt.axis([0,8,0,30]) x从0到8，y从0到30
plt.xlim(xmin=0, xmax=8)
plt.ylim(ymin=0, ymax=30)

# 坐标轴间隔
#plt.locator_params(nbins=10)
plt.locator_params('x', nbins=5)
plt.locator_params('y', nbins=6)

# 绘制图线并设置图线样式
plt.plot(x,y,":b")
plt.plot(x,np.sin(x)+4,"--c")
plt.plot(x,x*x/3+2, color=(0.1,0.2,0.3), marker='>')
plt.plot(np.cos(x)+4, color='#FF00FF', marker='s')

# 文本和公式
plt.text(5,20,"函数：y=2x+5")
plt.text(7,8,r"$y=\sin(x)+4$")

# 注释
plt.annotate("注释", xy=(3,11),xytext=(2,14),arrowprops={'facecolor':'r','shrink':0.05,'headwidth':15})


# 设置图例
plt.legend(['函数1','函数2','函数3','函数4'])

# 保存图片
plt.savefig('02-01.png')
plt.savefig('02-01.svg')

# 展示图片
plt.show()
```

**3. 复杂一点的例子**
- 包含子图、填充、双坐标等

```python
# 计算函数
x = np.arange(0,  3  * np.pi,  0.1) 

# 总标题
plt.suptitle('复杂一些的图')

# 调整子图参数
# top、bottom、right、top表示整个图距离边框距离
# wspace、hspace表示子图之间的间距
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=1, hspace=1)

# 绘制第一个图像
plt.subplot(2,  2,  1)
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'b') 
plt.title('正余弦')  

# 绘制第二个图像
plt.subplot(2,  2,  2) 
plt.plot(x, np.sin(x), 'r')
plt.twinx() #默认为0到1
plt.plot(x, np.log(x+1), 'b')
plt.title('双坐标')  

# 绘制第三个图像
plt.subplot(2,  2,  3)
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'b')
plt.fill(x, np.sin(x), 'r')
plt.fill(x, np.cos(x), 'b')
plt.title('正余弦填涂') 

# 绘制第四个图像
plt.subplot(2,  2,  4)
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'b')
plt.fill_between(x, np.sin(x), np.cos(x), where=np.sin(x)>np.cos(x), interpolate=True)
plt.title('正余弦填涂交叉位置')   

# 保存图片
plt.savefig('02-02.png')

# 展示图像
plt.show()
```
